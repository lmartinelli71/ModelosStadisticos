```python
from datetime import date, timedelta

#ajustar para obtener un mes
fecha_fin='2023-09-01' #str(date.today()-timedelta(10))
fecha_ini='2023-07-01' #str(date.today()-timedelta(41))

import yfinance as yf
from pandas_datareader import data as pdr
yf.pdr_override()

Datos_GOOG = pd.DataFrame(pdr.get_data_yahoo('GOOGL', start=fecha_ini, end=fecha_fin)).reset_index()
#Datos_GOOG=Datos_GOOG.assign(SYMBOL_YAHOO_FINANCE='GOOG')
Datos_GOOG
```

```python
#formato a Date
Datos_GOOG['Date'] = Datos_GOOG['Date'].dt.strftime('%Y-%m-%d')
```

```python
#Análisis de volatilidad

#calculo retornos porcentual diarios
Datos_GOOG['returns'] = Datos_GOOG['Close'].pct_change(1)
```

```python
Datos_GOOG['returns'].describe()
```

```python
fig = go.Figure()
fig.add_trace(go.Histogram(
    x=Datos_GOOG['returns'],
    #histnorm='probability', si se quiere normalizado
    marker_color='#EB89B5',
    autobinx=False,
    opacity=0.75
))

fig.update_layout(template='none')
```

```python
#retornos acumulados

Datos_GOOG['Cumulative Return'] = (1 + Datos_GOOG['returns']).cumprod()
```

```python
fig = go.Figure()
fig.add_trace(
    go.Scatter(x=Datos_GOOG['Date'],
               y=Datos_GOOG['Cumulative Return'],
               name="Cumulative Return",
               line=dict(color='#FECB52')
               ))
fig.update_layout(title_text="Retorno acumulado Google Julio -Agosto 2023")
fig.update_layout(dict(xaxis = dict( type='date', tickformat="%Y-%m-%d",range = ['2023-07-01','2023-09-1'],nticks=80,tickfont=dict(size=8),tickangle=45)))
fig.update_layout(template='none')
```
